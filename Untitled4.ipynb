#Chap.4 numpy套件(numeric python)
為什麼需要numpy?

numpy由c撰寫，速度快python 10~1000倍以上

numpy儲存結果較正確，python數字儲存不正確

numpy節省記憶體儲存空間

numpy的向量(矩陣計算)速度快

numpy的取值方便

```python
a_list=[1,2,3,4,5]
print(a_list*3)
```

```python
#numpy套件的ndarray型態,提供element-wise的操作
import numpy as np
a=np.array([1,2,3,4,5])
print(a*3)
```

```python
a=np.array([1,2,3,4,5])
print(type(a))
print(a.shape) #ndaary的形狀
print(a.dtype) #資料的型態
print(a.ndim) #維度
```

```python
import numpy as np
z=np.zeros(6)
print(z)
print(np.empty((2,3,4)))
print(np.arange(10))
```

#ndarray的進階操作

```python
my_array=np.arange(10)
print(my_array)
print(my_array[0])
print(my_array[0:5])
```

#2維資料取值

```python
my_2d_array=np.array([np.arange(0,5),np.arange(5,10)])
print(my_2d_array)
print(my_2d_array[1,:])
print(my_2d_array[:,1])
print(my_2d_array[1,1])
```

```python
my_array=np.arange(0,10)
print(my_array)
print(my_array.shape)
my_2d_array=my_array.reshape(2,5)
print(my_2d_array)
print((np.arange(0,100).reshape(10,10)))
```
